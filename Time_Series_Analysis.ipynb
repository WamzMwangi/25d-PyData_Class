```python
time_df['date.utc']=pd.to_datetime(time_df['date.utc'],format='mixed')
```

```python
time_df['date.utc'].dtype
```

```python
#Extract other datetime components from datetime object
time_df['Year']=time_df['date.utc'].dt.year
time_df.head()
```

```python
time_df['Month']=time_df['date.utc'].dt.month
time_df['Day']=time_df['date.utc'].dt.day
time_df['Day_name']=time_df['date.utc'].dt.day_name()
```

```python
time_df.head()
```

### Resampling our data based on time series components

```python
#asfreq()
time_df.set_index('date.utc',inplace=True)
time_df.head()
```

```python
day_df=time_df.asfreq(freq='D')
day_df
```

```python
paris_df=time_df.loc[time_df['city']=='Paris']
paris_df
```

```python
day_df=paris_df.asfreq(freq='D', method='ffill')
day_df
```

```python
#resample()
three_days=paris_df.resample('30S').sum(numeric_only=True)
three_days.head()
```

```python
three_days=paris_df.resample('30S').ffill()
```

```python
three_days.head()
```

```python
paris_df['3 day pollution']=paris_df['value'].rolling('3D').mean()
paris_df.head()
```
