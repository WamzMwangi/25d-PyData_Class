## Visualizations with Python Matplotlib
- Bar plots: Vertical, horizontal, stacked and grouped
- Customizations
- Scatter plots and color bars

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
drinks=pd.read_csv(r"C:\Users\Pula Advisors\Desktop\25d-PyData_Class\datasets\drinks.csv")
drinks.head()
```

```python
#Bar plots- visualize the counts of categorical features/variables
#we want to plot beer servings per country

plt.bar(drinks['country'], drinks['beer_servings'], color='blue', edgecolor='black')

plt.show()

```

```python
top_10=drinks.sort_values(by='beer_servings', ascending=False).head(10)
top_10
```

```python
plt.bar(top_10['country'], top_10['beer_servings'], color='lightblue', edgecolor='black')



plt.title('Beer Servings per person in Top 10')
plt.xticks(rotation=90)
plt.yticks(rotation=30)
plt.xlabel('Country')
plt.ylabel('Average beer serings per person')

plt.show()
```

```python
#advanced customization- adding annotations
plt.bar(top_10['country'], top_10['beer_servings'], color='lightblue', edgecolor='black')

plt.title('Beer Servings per person in Top 10')
plt.xticks(rotation=90)
plt.yticks(rotation=30)
plt.xlabel('Country')
plt.ylabel('Average beer serings per person')
plt.gca().axes.get_yaxis().set_visible(False)

#plt.text()- you have to define for it an x and y coordinate
for x, y in enumerate (top_10['beer_servings']):
    plt.text(x, y+5, str(y), ha='center')

plt.show()
```

### Grouped Bar chart

```python
plt.figure(figsize=(10,6))


bar_width= 0.5
x=np.arange(len(top_10['country']))

plt.bar(x-bar_width/2, top_10['beer_servings'],bar_width, label='Beer', color='blue')
plt.bar(x+bar_width/2, top_10['spirit_servings'],bar_width, label='Wine', color='gray')


plt.xticks(x,top_10['country'],rotation=90)
plt.title('Beer & Spirit Servings per person in top 10 countries')

for x, y in enumerate (top_10['beer_servings']):
    plt.text(x, y+5, str(y), ha='center',color='b')

for x, y in enumerate (top_10['spirit_servings']):
    plt.text(x, y+5, str(y), ha='left',color='gray')



plt.show()
```
