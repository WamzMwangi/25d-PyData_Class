### Masks, Comparisons & Boolean operators in Numpy

```python
import numpy as np

weights=np.array([52,43,61,30,72,51,28,64])
weights
```

```python
obese_mask=weights>60
obese_mask
```

```python
obese_patients=weights[obese_mask]
obese_patients
```

```python
not_obese_mask= ~(weights<60)
not_obese_mask
```

```python
not_obese_patients=weights[not_obese_mask]
not_obese_patients
```

```python
moderate_mask=(weights>=40) & (weights<=60)
moderate_mask
```

## Grouping & Aggregation

```python
#groupby function
# split-apply- combine
#df.groupby('split')['apply'].agg()
sales_df.groupby('Brands')['Selling Price'].mean().sort_values(ascending=False)
```

```python
# Which brand has been given the highest average discount
sales_df.groupby('Brands')['Discount'].mean().idxmax()
```

```python
# which brand did sell the most of 
sales_df.groupby('Brands')['Selling Price'].count().idxmax()
```

```python
# which brand has the highest number of models
sales_df.groupby('Brands')['Models'].nunique().sort_values(ascending=False)
```

```python
sales_df.groupby('Brands')['Selling Price'].agg(lambda x: x.max()- x.min())
```
