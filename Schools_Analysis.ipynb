```python
#import all necessary libarries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.bar(df['country'],df['beer_servings'],color='purple',edgecolor='black')

plt.xlabel('Country')
plt.ylabel('Beer servings per person')
plt.title('Average beer servings per country')
plt.show()
```

```python
top10_beer = df.sort_values(by='beer_servings', ascending=False).head(10)
top10_beer
```

```python
plt.bar(top10_beer['country'],top10_beer['beer_servings'],color='purple',edgecolor='black')

plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Beer servings per person')
#plt.gca().axes.get_yaxis().set_visible(False)
plt.title('Average beer servings per country')


for x, y in enumerate(top10_beer['beer_servings']):
    plt.text(x,y+5, str(y),ha='center')
plt.show()
```

Grouped Bar Charts

```python
top10_beer
```

```python
bar_width=0.5
x=np.arange(len(top10_beer['country']))

plt.bar(x-bar_width/2,top10_beer['beer_servings'],bar_width,label='Boy schools',color='blue')
plt.bar(x+bar_width/2,top10_beer['spirit_servings'],bar_width,label='Girl schools',color='green')

plt.ylabel('No.of schools')
plt.xlabel('Counties')
plt.title('Comparison of Boy & Girl Schools across Counties')
plt.xticks(x,top10_beer['country'],rotation=90)
plt.legend()

for x, y in enumerate(top10_beer['beer_servings']):
    plt.text(x,y+5, str(y),ha='center')
for x, y in enumerate(top10_beer['spirit_servings']):
    plt.text(x,y+5, str(y),ha='center')

plt.show()
```

```python
plt.bar(top10_beer['country'], top10_beer['beer_servings'], color='r')
plt.bar(top10_beer['country'], top10_beer['spirit_servings'], bottom=top10_beer['beer_servings'], color='b')
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(8,6))

axs[0].bar(top10_beer['country'], top10_beer['beer_servings'], color='r')
axs[1].bar(top10_beer['country'], top10_beer['spirit_servings'], color='b')


plt.show()
```
